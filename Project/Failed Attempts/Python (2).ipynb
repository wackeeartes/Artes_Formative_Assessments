```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Set up the WebDriver
service = Service(ChromeDriverManager().install())
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# Open the Google Maps URL
resto_url = "https://www.google.com/maps/place/Ramen+Nagi-UP+Town+QC/@14.6499754,121.0753206,17z/data=!3m1!4b1!4m6!3m5!1s0x3397b76319dda73f:0xf8cefb505e213387!8m2!3d14.6499754!4d121.0753206!16s%2Fg%2F11bw2gjfpm?authuser=0&hl=en&entry=ttu"
driver.get(resto_url)

# Allow time for the page to load
driver.implicitly_wait(10)

# Find the restaurant name
name_element = driver.find_element(By.CLASS_NAME, "DUwDvf.lfPIob")
name = name_element.text

time.sleep(1.5)

#Find rating
rating_element = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[2]/div/div[1]/div[2]/span[1]/span[1]')
rating = rating_element.text

#find address
address_element = driver.find_element(By.CLASS_NAME, "rogA2c ")
address = address_element.text

#Open
open_element = driver.find_element(By.CLASS_NAME, "ZDu9vd")
open = open_element.text

#phone number
phone_element = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[9]/div[7]/button/div/div[2]')
phone = phone_element.text



print(str("Restaurant: ") + str(name))
print(str("Rating: ") + str(rating))
print(str("Address: ") + str(address))
print(open)
print(str("Phone Number: ") + str(phone))

# Close the WebDriver
driver.quit()
```
