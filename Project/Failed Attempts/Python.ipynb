```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Set up the WebDriver
service = Service(ChromeDriverManager().install())
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

for link in resto_links:
# Open the Google Maps URL
    resto_url = link
    driver.get(resto_url)
    
    # Allow time for the page to load
    driver.implicitly_wait(10)

    try:
        # Find the restaurant name
        name_element = driver.find_element(By.CLASS_NAME, "DUwDvf.lfPIob")
        name = name_element.text
    except NoSuchElementException:
        name = "N/A"
    
    time.sleep(1.5)
    
    try:
        # Find rating
        rating_element = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[2]/div/div[1]/div[2]/span[1]/span[1]')
        rating = rating_element.text
    except NoSuchElementException:
        rating = "N/A"
    
    try:
        # Find address
        address_element = driver.find_element(By.CLASS_NAME, "rogA2c")
        address = address_element.text
    except NoSuchElementException:
        address = "N/A"
    
    try:
        # Find open status
        open_element = driver.find_element(By.CLASS_NAME, "o0Svhf")
        open_status = open_element.text
    except NoSuchElementException:
        open_status = "N/A"
    
    try:
        # Find phone number
        phone_element = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[9]/div[7]/button/div/div[2]')
        phone = phone_element.text
    except NoSuchElementException:
        phone = "N/A"

    
    print(str("Restaurant: ") + str(name))
    print(str("Rating: ") + str(rating))
    print(str("Address: ") + str(address))
    print(str(open))
    print(str("Phone Number: ") + str(phone))
    print("------------------------------------")

# Close the WebDriver
driver.quit()
```
