```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# Navigate to Google Maps
url = 'https://www.google.com/maps'
driver.get(url)
time.sleep(1)

# Zoom in halfway
try:
    # Locate the zoom control slider element
    zoom_slider = driver.find_element(By.XPATH, '//*[@id="widget-zoom-slider"]')
    
    # Get the slider's width
    slider_width = zoom_slider.size['width']
    
    # Calculate the position to click for zooming halfway
    half_width = slider_width / 2
    action = webdriver.ActionChains(driver)
    action.click_and_hold(zoom_slider).move_by_offset(half_width, 0).release().perform()
    time.sleep(1)  # Optional: Adjust this sleep time as needed
except Exception as e:
    print(f"Failed to zoom in: {e}")

# Click the "Restaurants" button
restaurants_button = driver.find_element(By.XPATH, '//*[@id="searchbox-searchbutton"]/span')
restaurants_button.click()

# Clear the search bar
search_input = driver.find_element(By.XPATH, "//input[@name='q' and @id='searchboxinput']")
search_input.clear()

# Prompt user for input
city = input("Enter the city: ")
landmark = input("Enter the landmark: ")
cuisine = input("Enter the cuisine (e.g., Japanese): ")
specialty = input("Enter the specialty (e.g., Chicken Karaage): ")
rating = float(input("Enter the minimum rating (0.0-5.0): "))
restrictions = input("Enter dietary restrictions: ")
features = input("Enter features: ")
service_options = input("Enter service options (Delivery, Takeout, Dine-in): ")

# Search query construction
search_query = f"{cuisine} restaurant near {landmark}, {city} with good {specialty}, {restrictions} {features} {service_options}"

# Enter search query
search_input.send_keys(search_query)

# Click search button
driver.find_element(By.XPATH, '//*[@id="searchbox-searchbutton"]/span').click()

time.sleep(5)

# Apply filters based on rating
if rating >= 2:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[2]').click()
if rating >= 2.5:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[3]').click()
if rating >= 3:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[4]').click()
if rating >= 3.5:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[5]').click()
if rating >= 4:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[6]').click()
if rating >= 4.5:
    driver.find_element(By.XPATH, '//*[@id="omnibox-singlebox"]/div[2]/div[1]/div[1]/div[1]/button').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="action-menu"]/div[7]').click()

time.sleep(5)

# Extract and print the first 3 restaurant options
restaurants = driver.find_elements(By.XPATH, '//*[@class="section-result"]')[:3]
for restaurant in restaurants:
    try:
        name = restaurant.find_element(By.XPATH, './/h3').text
    except:
        name = "N/A"
    try:
        rating = restaurant.find_element(By.XPATH, './/span[@aria-label]').get_attribute('aria-label')
    except:
        rating = "N/A"
    try:
        address = restaurant.find_element(By.XPATH, './/span[@class="section-result-location"]').text
    except:
        address = "N/A"
    try:
        open_status = restaurant.find_element(By.XPATH, './/span[contains(text(), "Open") or contains(text(), "Closed")]').text
    except:
        open_status = "N/A"
    try:
        phone = restaurant.find_element(By.XPATH, './/span[@class="section-result-phone-number"]').text
    except:
        phone = "N/A"

    print(f"Restaurant: {name}")
    print(f"Rating: {rating}")
    print(f"Address: {address}")
    print(f"Open Status: {open_status}")
    print(f"Phone Number: {phone}")
    print("-" * 30)

time.sleep(10)
driver.quit()
```
